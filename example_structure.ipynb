# Strukturen Beispiele aus RNALoops

```python
from rnaloops.data_explorer.import_helper import *
df = get_prepared_df()
```

```python
# Beispiele von home_structure mit 5 multiploops:
home_grouped = df.groupby('home_structure').count().sort_values(by='loop_type')
indices = home_grouped[home_grouped.loop_type == 5].index.values
bsp = df[df.home_structure.isin(indices)].sort_values(by='home_structure')
bsp = bsp[25:30]
show_structure(indices=bsp.index.values, web=True)
```

```python
# Beispiele von Strukturen die sehr häufig vorkommen:
dublicated_sequences = df.groupby('whole_sequence').count()
dublicated_sequences.sort_values(by='loop_type', ascending=False).head()
```

```python
# Und Visualisierung zugehöriger Winkelverteilung:
most_freq_seq = df.way5[df.way5.whole_sequence == 'CG — UAAAG — CUAA — CG — CG']
set_attrs(most_freq_seq)
_ = most_freq_seq.euler.hist(figsize=(12, 6), grid=False, bins=100)
plt.tight_layout()
```

```python
# Anzahl Strukturen ohne Pseudoknoten (vermutlich):
without_pseudo = df[~df.db_notation.str.contains("\}|\]")]
'Structures without pseudo knots:', len(without_pseudo.index)
```

```python
# Kleinste home_structure die auch in RNAStrands gefunden wurde:
smallest_home_structure = df[df.home_structure=='1nyi']
show_structure(indices=smallest_home_structure.index, web=True)
```
